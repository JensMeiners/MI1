```python
# imports
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# preparing data
def f_(x):
    return -np.sin(5*np.pi*x) + 2*np.power(2*x - 1, 3)

def gen_data(mean=0,sigma=1,start=0,end=1.01,steps=0.02):
    x = np.arange(start, end, steps)
    y = f_(x) + np.random.normal(mean,sigma,x.shape)
    return x,y

def normalize_data((x,y)):
    # zero mean and range from -1 to 1
    x -= x.mean()
    x = (x / x.max()) 
    y -= y.mean()
    y = (y / y.max()) 
    return x,y
    
    
```

```python
x,y = gen_data()
plt.plot(x,y)
plt.show()
```

```python
def poly(x,y,order):
    c = np.polyfit(x,y,order)
    return np.poly1d(c)
```

```python
## test monomial function
# generate data
x,y = normalize_data(gen_data())
plt.scatter(x,y)
#plt.plot(x,y)

# find coefficients for order 3
p = poly(x,y,3)

# plot the result
p_x = np.linspace(-1,1,num=20)
plt.plot(p_x, p(p_x))
plt.show()

print 'x.max',x.max(),'x.min',x.min()

# show coefficiants used
print 'coefficients (starting with highest power):', p.c
```

```python
# generate 40 different samples
N = 40
raw_data = []
for i in range(N):
    x,y = normalize_data(gen_data())
    raw_data.append((x,y))
    
orders = [1,5,9,14]
nets = []
datasets = []
for o in orders:
    _in, _out = o + 1, 1
    net = buildNetwork(_in, _out, bias=False)
    nets.append(net)
    ds = SupervisedDataSet(_in, _out)
    datasets.append(ds)
    for i in range(N):
        sample_x, sample_y = cn_input(raw_data[i][0], raw_data[i][1], o)
        normalize_sample(sample_x, sample_y)
        decorrelate_sample(sample_x, sample_y)
        #print sample_x.shape,sample_y.shape
        #ds.addSample(sample_x,sample_y)
        for i,sample in enumerate(sample_x):
            ds.addSample(sample, sample_y[i])
    trainer = BackpropTrainer(net, ds)
    trainer.trainUntilConvergence(maxEpochs=10)
```

```python
import numpy as np
from numpy import genfromtxt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from pybrain.supervised.trainers import RPropMinusTrainer
from pybrain.datasets import SupervisedDataSet
from pybrain.tools.shortcuts import buildNetwork
from pybrain.tools.validation import ModuleValidator, CrossValidator
from sklearn import preprocessing

data = genfromtxt('TrainingRidge.csv', delimiter=',', skip_header=1)
data = preprocessing.scale(data)
#print data
trndata = data[:,0:2]
trudata = data[:,2]


trndataExp = np.empty((10,200,2))

for i in range(10):
    trndataExp[i]=trndata**i
    
   
print trndataExp.shape

ds = SupervisedDataSet(2, 1)
for i in range(10):
    for j in range(10-i):
        #ds.addSample(data[i,0:2], data[i,2])
        for n in range(200):
            ds.addSample((trndataExp[i,n,0],trndataExp[j,n,1]), data[n,2])
            
            
print ds

#3d plot of the data
#fig = plt.figure()
#ax = fig.add_subplot(111, projection='3d')
#ax.scatter(data[:,0], data[:,1], data[:,2], c='r', marker='o')
#ax.set_xlabel('X Label')
#ax.set_ylabel('Y Label')
#ax.set_zlabel('Z Label')
#plt.show()

```

```python
from pybrain.tools.validation import Validator




for weightdecay in (0.01,0.02,0.05,0.1,0.2,0.5,1,1.5,3,5,10):
    print weightdecay

    
    
    #cross validation
    mse = 0
    folds = 10
    for i in range(folds):
        net = buildNetwork(2,1)
        trainer = RPropMinusTrainer(net, dataset=ds, verbose=False, weightdecay=weightdecay)
        dstrn,dsval = ds.splitWithProportion(proportion=0.1)
        trainer.trainUntilConvergence(dataset=dstrn,maxEpochs=20)
        output = net.activateOnDataset(dsval)

        validator = Validator()
        mse +=validator.MSE(output,dsval['target'])
        #print i
    print 'MSE',mse/10
    print '--------------'

print 'DONE'
```
